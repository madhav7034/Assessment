```python
#8

df = df.dropna(subset=['Revenue (Millions)', 'Metascore'])
df
```

```python
#11
avg_rating_by_director = df.groupby('Director')['Rating'].mean().sort_values(ascending=False)
print("Average Rating by Director:\n")
avg_rating_by_director
```

```python
#17

conditions=[df['Rating'] < 6 ,(df['Rating']>=6) & (df['Rating']<=7.5) ,df['Rating'] >=7.5 ]
values =["poor",'average','excellent']
df['movie_rating']= np.select(conditions,values,default='UNKNOWN')
df.head(10)
```

```python
#18

def calculate_high_score(row):

    return row['Rating'] * 10 + row['Revenue (Millions)']
df['High_Score'] = df.apply(calculate_high_score, axis=1)
print(df[['Movie Title', 'Rating', 'Revenue (Millions)', 'High_Score']].head())
```
