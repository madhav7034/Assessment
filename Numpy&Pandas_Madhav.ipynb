```python
#1

import numpy as np

arr = np.arange(10, 50)
reshaped = arr.reshape(4, 10)
last_col = reshaped[:, -1]

print("1. Original Array:\n", arr)
print('\n')
print("Reshaped (4x10):\n", reshaped)
print('\n')
print("Last Column:\n", last_col)
```

```python
#2


mat1 = np.full((3, 3), 5)
mat2 = np.arange(1, 10).reshape(3, 3)

added = mat1 + mat2
multiplied = mat1 * mat2
transposed = multiplied.T

print("\n2. Matrix 1 (All 5s):\n", mat1)
print("Matrix 2 (1 to 9):\n", mat2)
print("Added:\n", added)
print("Element-wise Multiplied:\n", multiplied)
print("Transpose of Result:\n", transposed)
```

```python
#3

rand_arr = np.random.randint(10, 101, 15)
replaced_arr = np.where(rand_arr > 50, -1, rand_arr)
count_replaced = np.sum(rand_arr > 50)

print("\n3. Original Random Array:\n", rand_arr)
print("After Replacing >50 with -1:\n", replaced_arr)
print("Count of Values Replaced:", count_replaced)
```

```python
#4

array_3 = np.random.random((5,4))
row_sums = np.sum(array_3, 1)
max_sum_ind = np.argmax(row_sums)

def normalize(arr):
  row_min = np.min(arr, axis=-1, keepdims=True)
  row_max = np.max(arr, axis=-1, keepdims=True)
  norm = (arr - row_min) / (row_max - row_min)
  return np.round(norm, 2)



normalized = normalize(array_3)

print(f'5X4 row :\n{array_3}\n')
print(f'Row with maximum sum:\n{array_3[max_sum_ind]}\n')
print(f'Normalized array:\n{normalized}')
```

```python
#5

arr = np.array([1, 3, 5, 3, 7, 9, 1, 5, 5])
unique = np.unique(arr)

counts = np.bincount(arr)
duplicates = np.where(counts > 1)[0]

print("Unique Elements:\n", unique)
print("Duplicates:\n", duplicates)
```

```python
#6

arr = np.arange(1, 37).reshape(6, 6)
submatrix = arr[1:5, 1:5]
even_numbers = submatrix[submatrix % 2 == 0]

print("\n6. Original 6x6 Array:\n", arr)
print("Middle 4x4 Submatrix:\n", submatrix)
print("Even Numbers in Submatrix:\n", even_numbers)
```

```python
#7

arr = np.arange(12).reshape(3, 4)
reshaped = arr.reshape(2, 6)
flattened = reshaped.flatten()
reshaped_again = flattened.reshape(4, 3)

print("\n7. Original 3x4 Array:\n", arr)
print("Reshaped to 2x6:\n", reshaped)
print("Flattened to 1D:\n", flattened)
print("Reshaped to 4x3:\n", reshaped_again)
```

```python
#8



a = np.array([[1, 2], [3, 4]])
row_added = np.vstack([a, [[5, 6]]])
column_added = np.hstack([row_added, [[7], [8], [9]]])

print(f'array a:\n{a}\n')
print(f'array after appending row:\n{row_added}\n')
print(f'array after appending row and column:\n{column_added}')
```

```python
#9


a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.array([7, 8, 9])
d = np.concatenate([a, b, c])
e = d.reshape((3, 3))
f = np.diag(e)

print(f'Joined array d:\n{d}\n')
print(f'Reshaped array:\n{e}\n')
print(f'Extracted elements:\n{f}')
```

```python
#10

arr = np.arange(1, 21)
filtered = arr[(arr % 3 == 0) | (arr % 5 == 0)]
count = filtered.size

print("\n10. Original Array:\n", arr)
print("Values Divisible by 3 or 5:\n", filtered)
print("Count:", count)
```

```python
#1

data=sns.load_dataset("titanic")
```

```python
#2

data.isna().sum()
```

```python
#3

percentage=data.groupby("sex")["survived"].mean()*100
percentage.round(2)
```

```python
#4


age=data['age']
avr=age.mean()
med=age.median()
std_age=age.std()
print(avr)
print(med)
print(std_age)
```

```python
#5

pclass=data['pclass']
pclass.value_counts()
```

```python
#6

survived_female = data[(data['sex'] == 'female') & (data['age'] < 18) & (data['survived'] == 1)]
display(survived_female)
```

```python
#7

age.fillna(med,inplace=True)
age
```

```python
#8


conditions = [data['age'] < 18, (data['age'] >= 18) & (data['age'] < 60), data['age']>60]
values = ['Child','Adult', 'Senior']
data['age_group'] = np.select(conditions, values, default='Unknown')
data.head(10)
```

```python
#9

data = data.sort_values("fare",ascending=False)
data.head()
```

```python
#10

data.groupby(['pclass','sex'])["fare"].mean()
```
