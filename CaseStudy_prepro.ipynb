```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
for i in num_cols:
  plt.figure(figsize=(5,5))
  plt.boxplot(new_house[i])
  plt.title(i)
```

```python
q1 =np.quantile(new_house['No of Bathrooms'],0.25)
q3 =np.quantile(new_house['No of Bathrooms'],0.75)
iqr = q3 - q1
upper_bound = q3+(1.5*iqr)
lower_bound = q1-(1.5*iqr)
print(upper_bound)
print(lower_bound)
print(iqr)
```

```python
bath_out =[]
for i in new_house['No of Bathrooms']:
  if i > upper_bound or i < lower_bound:
    bath_out.append(i)
```

```python
print(bath_out)
```

```python
new_house['No of Bathrooms'] = new_house['No of Bathrooms'].clip(lower=lower_bound,upper=upper_bound)
```
